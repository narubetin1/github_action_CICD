```python
import duckdb
import pandas as pd

# เชื่อมต่อ DuckDB (ไม่ต้องใส่ไฟล์)
con = duckdb.connect()

# ติดตั้งและโหลด SQLite extension
con.execute("INSTALL sqlite; LOAD sqlite;")

# แนบฐานข้อมูล SQLite (airbnb.db)
con.execute("ATTACH 'airbnb.db' AS sqlite_db (TYPE sqlite);")

# ตรวจสอบรายชื่อตาราง
#tables = con.execute("SHOW TABLES FROM airbnb_listings").fetchall()
#print("Tables:", tables)

# สมมติว่ามีตารางชื่อ 'listings' ให้ดึงข้อมูล
df = con.execute("SELECT * FROM airbnb.airbnb_listings").fetchdf()

print(df.head())
```
